### Chunkining

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
sample ="""The word brown dog went running through the mud."""


tokanize_text = nltk.word_tokenize(sample)
print(tokanize_text)
```

###### noun phraces

```python
pos_text = nltk.pos_tag(tokanize_text)

pos_text
```

```python
grammer = (r'''
NP: {<DT>?<JJ>*<NN>*}''')
```

```python
chunkParser = nltk.RegexpParser(grammer)
```

```python
tree = chunkParser.parse(pos_text)
```

```python
print(tree)
```

```python
tree.draw()
```

```python
import nltk
from nltk import pos_tag, word_tokenize

# Example sentence
sample = "The dark cloud coverd the sky ."

# Tokenize the sentence
tokens = word_tokenize(sample)

# Perform part-of-speech tagging
pos = pos_tag(tokens)

# Display the tagged tokens
print(pos)
```

#### Chinking

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
sample = """The crazy brown dog went running through the mud."""

pos_text = nltk.pos_tag(nltk.word_tokenize(sample))

pos_text
```

```python
grammer = (r"""Chunk: {<.*>}
}<VB.?|IN|DT|TO>+{ """)  #Sequence of tokens for chinking
```

```python
chunkParser = nltk.RegexpParser(grammer)
```

```python
tree = chunkParser.parse(pos_text)
```

```python
print(tree)
```

```python
tree.draw()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import RegexpParser
```

```python
sample = "Marry saw a cat sitting on the mat"
```

```python
pos = pos_tag(word_tokenize(sample))

pos
```

```python
regex = r"""
              NP: {<DT>?<JJ>*<NN.*>}   #to extract noun phraces
              p: {<IN>}                #to exract prepositions
              V: {<V.*>}               #to extract verbs
              PP: {<P> <NP>}           #to extract prepositioinal phraces
              VP: {<V> <NP|PP>*}       #to extract verb phraces
        """
```

```python
parser = RegexpParser(regex)
```

```python
tree = parser.parse(pos)
```

```python
print(tree)
```

```python
tree.draw()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import RegexpParser
```

```python
sample ="the little yellow dog barked at the cat"
```

```python
pos = pos_tag(word_tokenize(sample))
```

```python
pos
```

```python
regex = r"""
              NP: {<.*>+}    # chunk everything
              }<VBD|IN>+{    #chink sequences of VBD and IN
        """
```

```python
parser = RegexpParser(regex)
```

```python
tree = parser.parse(pos)
```

```python
print(tree)
```

```python
tree.draw()
```
